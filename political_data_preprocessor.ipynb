```python
import csv 
import numpy as np
from sklearn.utils import shuffle
```

```python
#TRAINING DATA
```

```python
pol1, pol2, pol3 = list(np.load('./Noyce/data/political_comments1.npy')), list(np.load('./Noyce/data/political_comments2.npy')), list(np.load('./Noyce/data/political_comments3.npy'))
apol1, apol2, apol3 = list(np.load('./Noyce/data/apolitical_comments1.npy')), list(np.load('./Noyce/data/apolitical_comments2.npy')), list(np.load('./Noyce/data/apolitical_comments3.npy'))
```

```python
pol = pol1 + pol2 + pol3
apol = apol1 + apol2 + apol3

print(len(pol), len(apol))
```

```python
pol_class_id = [1]*len(pol)
apol_class_id = [0]*len(apol)
```

```python
text = pol + apol
class_id = pol_class_id + apol_class_id

class_id, text = shuffle(class_id, text, random_state=42)
```

```python
print(len(class_id), len(text))
```

```python
with open('./Noyce/data/anshuman_political/political_train.csv', 'w+', newline='', encoding='utf-8') as csvfile:
    csvwriter = csv.writer(csvfile, delimiter=',')
    csvwriter.writerow(['class_id', 'text'])
    csvwriter.writerows(zip(class_id, text))
```

```python
i = 0
class_id = []
text = []
with open('./Noyce/data/political_text/political_train.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if i != 0:
            class_id.append(int(row[0]))
            text.append(str(row[1]))
        i += 1

i = 0
with open('./Noyce/data/political_text/political_test.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        if i != 0:
            class_id.append(int(row[0]))
            text.append(str(row[1]))
        i += 1
```
